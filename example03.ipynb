```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from scipy import linalg


w0=0.1;
dt=0.1;

F=np.matrix([[0, 1], [-w0*w0, 0]])
Phi=linalg.expm(F*dt)
print(Phi)

#initial vales corresponding to y(t)=cos(t)
x=np.array([1,0])
N=int(400/dt)
x_solution=np.zeros((2, N+1))
x_solution[:,0]=np.asarray(x)
for i in range(1,N+1):
    x=Phi@x
    x_solution[:,i]=x

t=np.linspace(0, 400, num=(N+1))

```

Now plot and compare with the correct solution

```python
plt.figure(1,figsize=(12, 9))
plt.plot(t, x_solution[0,:], 'r-')
plt.plot(t,np.cos(w0*t))
plt.show()
```

Hard to see the difference. So compute it

```python
plt.figure(1,figsize=(12, 9))
plt.plot(t, x_solution[0,:]-np.cos(w0*t), 'r-')
```

Now we are ready for the next example, the damped osciallotor

```python
damp=0.1
F=np.matrix([[0, 1], [-w0*w0, -damp*w0]])
Phi=linalg.expm(F*dt)
print(Phi)
```
